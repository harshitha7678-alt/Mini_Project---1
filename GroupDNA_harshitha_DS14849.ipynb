```python
import re
import numpy as np
from collections import Counter, defaultdict
filename = "hostel_bois.txt"

with open(filename, "r", encoding="utf-8") as f:
    chats = f.readlines()

print("Total Raw Lines :", len(chats))

users = []
messages = []
hours = []
days = []
all_words = []

pattern = r"(\d+/\d+/\d+),\s(\d+):(\d+)\s-\s([^:]+):\s(.*)"

for line in chats:

    match = re.match(pattern, line)

    if match:

        days.append(match.group(1))
        hours.append(int(match.group(2)))
        users.append(match.group(4))
        messages.append(match.group(5))

user_count = Counter(users)


stopwords = {
    "the","is","a","an","to","of",
    "and","you","are","i","me",
    "my","we","our","this","that"
}

for msg in messages:

    for w in msg.lower().split():

        w = re.sub(r'[^a-z]', '', w)

        if len(w) > 2 and w not in stopwords:
            all_words.append(w)

word_count = Counter(all_words)



print("\nMessages :", len(messages))
print("Members :", len(set(users)))
print("Days :", len(set(days)))

print("\nMessage per person:")
for u, c in user_count.items():
    print(u, ":", c)

print("\nTop Words:")
for w, c in word_count.most_common(5):
    print(w, ":", c)
```

**Feature 1 - The Chat Parser**

```python
with open("hostel_bois.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Store parsed messages
messages = []

# Counters
system_messages = 0
media_messages = 0
deleted_messages = 0

participants = set()

# Parse each line
for line in lines:

    line = line.strip()

    # Ignore empty lines
    if line == "":
        continue

    # Every valid WhatsApp line contains " - "
    if " - " not in line:
        continue

    # Split timestamp and remaining text
    timestamp, remaining = line.split(" - ", 1)

    # Skip system messages
    if ": " not in remaining:
        system_messages += 1
        continue

    # Split sender and message
    sender, message = remaining.split(": ", 1)

    # Skip media messages
    if message.strip() == "<Media omitted>":
        media_messages += 1
        continue

    # Skip deleted messages
    if message.strip() in [
        "This message was deleted",
        "You deleted this message"
    ]:
        deleted_messages += 1
        continue

    # Store parsed message
    messages.append({
        "timestamp": timestamp,
        "sender": sender,
        "message": message
    })

    participants.add(sender)

# ==========================================================
# OUTPUT
# ==========================================================

print("=" * 60)
print("FEATURE 1 : CHAT PARSER")
print("=" * 60)

print("Total Parsed Messages :", len(messages))
print("Total Participants    :", len(participants))
print("System Messages       :", system_messages)
print("Media Messages        :", media_messages)
print("Deleted Messages      :", deleted_messages)

print("\n")
```
